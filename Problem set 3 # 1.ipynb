```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
data = np.genfromtxt('sdssdr7eq.csv', skip_header=2, delimiter=(','))
```

```python
ra=[datum[0] for datum in data if datum[2] > 0] #import only galaxies with positive redshifts
z  =[datum[2] for datum in data if datum[2] > 0]
Z   = np.array([z]) #convert lists to numpy arrays
rad = np.array([ra])
rad = (rad*(np.pi/180)) - (11*np.pi)/20 #convert right ascension to radians, rotate by pi/2
d = (299792.458*Z)/72 #convert redshifts into distances in Megaparsecs, using c*z = Ho*d
ra = np.ndarray.tolist(rad) #convert right ascension in radians back to list for better performance
np.ndarray.tolist(d) #repeat process for remaining arrays
z = np.ndarray.tolist(Z)
```

```python
plt.figure(figsize = (25,25))
fig = plt.subplot(111, projection = 'polar')
fig.plot(ra, d, 'm.')
fig.set_rmax(500)
plt.title('galactic positions, plotted as a function of right ascension angle and distance', loc=('left'))
#ylabel = plt.ylabel('distane (Mpc)')
fig.set_xticklabels(['99\N{DEGREE SIGN}', '144\N{DEGREE SIGN}', '189\N{DEGREE SIGN}','234\N{DEGREE SIGN}',
                     '279\N{DEGREE SIGN}', '324\N{DEGREE SIGN}', '9\N{DEGREE SIGN}', '54\N{DEGREE SIGN}'])
fig.set_rlabel_position(30)
plt.savefig('universe.png')
plt.show()
```
