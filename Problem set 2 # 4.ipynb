```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
SDSS = np.genfromtxt('sdss.csv', skip_header=3, delimiter=(','))
```

```python
z=[datum[2] for datum in SDSS if datum[2] > 0]
r=[datum[4] for datum in SDSS if datum[2] > 0]
g=[datum[3] for datum in SDSS if datum[2] > 0]
Z=np.array([z])
R=np.array([r])
G=np.array([g])
R=np.array([r])
gMinusr = G-R
ZStar=(4.164*10**8)*Z
absoluteR=R-5*np.log10(ZStar)
```

```python
plt.figure(figsize = (15,10))
fig = plt.plot(Z,absoluteR, 'm.')
plt.axis([-0.025,0.225,-12,-26])
plt.grid(True)
plt.title('absolute r-magnitude as a function of redshift (z)')
xlabel = plt.xlabel('redshift (z)') 
ylabel = plt.ylabel('absolute r-magnitude')
plt.savefig('chart1.png')
```

```python
plt.figure(figsize = (15,10))
fig = plt.plot(absoluteR,gMinusr, 'm.')
plt.axis([-14,-25,0,1.5])
plt.grid(True)
plt.title('g-r as a function of absolute r-magnitude')
xlabel = plt.xlabel('absolute r-magnitude') 
ylabel = plt.ylabel('g-r')
plt.savefig('chart2.png')
```
